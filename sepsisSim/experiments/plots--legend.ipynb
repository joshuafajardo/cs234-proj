```python
%config InlineBackend.figure_formats = ['svg']
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.sans-serif'] = ['Arial']
```

```python
fig, ax = plt.subplots(figsize=(0.1,0.1))
plt.legend(
    [matplotlib.lines.Line2D([], [], c='tab:blue', lw=1, marker='s', ms=3), 
     matplotlib.lines.Line2D([], [], c='w'),
     matplotlib.lines.Line2D([], [], c='gray', ls=(0,(0.5,0.5))), 
     matplotlib.lines.Line2D([], [], c='k', marker='*', linestyle='none'),
     matplotlib.lines.Line2D([], [], c='r', ls=(0,(1,1)), marker='o', ms=3), 
     matplotlib.lines.Line2D([], [], c='g', ls='--', marker='D', ms=3), 
     ],
    ['C*-PDIS, noisy annot.', 
     '',
     'PDIS (baseline)', 
     'ideal case', 
     'C-PDIS , w/o imputation', 
     'C*-PDIS, w/ imputation'],
    loc='center', bbox_to_anchor=(0.5,1.0),
    ncol=3, handlelength=1.333, handletextpad=0.6, columnspacing=1,
)
ax.axis(False)
plt.savefig('fig/annotation_legend.pdf', bbox_inches='tight')
plt.show()
```
